**Student Marks problem**
Problem (Real Life)

You are given marks of 5 students in 3 subjects:

Math

Physics

CS

You must:

Store data using NumPy

Analyze performance

Filter students

Modify data

Use slicing, masking, aggregation, broadcasting



**Data storing**

```python
import numpy as np
marks=np.array([
    [85, 78, 92],   # Student 1
    [60, 55, 65],   # Student 2
    [90, 88, 95],   # Student 3
    [45, 50, 40],   # Student 4
    [72, 68, 70]    # Student 5
])
"""
         CS  Math  Physics
student1 [85, 78,   92],
student2 [60, 55,   65],
student3 [90, 88,   95],
student4 [45, 50,   40],
student5 [72, 68,   70]

"""
print(marks)
```

**Basic Info**

```python
print(marks.shape)#Students,Subjects
print(marks.size)#Total elements
print(marks.dtype)#Array data type
```

**Average marks per student**

```python
avg_students=np.mean(marks,axis=1)
print(avg_students)
for i,avg in enumerate(avg_students):
  print(f"Average marks of student{i+1}:{avg}")
```

**Average marks per subject**

```python
avg_students1=np.mean(marks,axis=0)
print(avg_students1)
for i,avg in enumerate(avg_students1):
  if i==0:
    print(f"Average of marks in CS:{avg}")
  elif i==1:
    print(f"Average of marks in Math:{avg}")
  else:
    print(f"Average of marks in Physics:{avg}")
```

**Top Students**

```python
top_students=avg_students>=80#Boolean Masking
print(top_students)
print(marks[top_students])
```

```python
print(avg_students)
```

**Weaker Students**

```python
weaker_student=avg_students<=60#Boolean Masking
print(weaker_student)
print(marks[weaker_student])
```

**Grace Marks**

```python
marks=marks+5
print(marks)
```

```python
marks[marks>100]=100
print(marks)
```




**Analyzing CS marks with slicing**



```python
cs_marks=marks[:,0]
print(cs_marks)
```

**Math marks of first 3 students**

```python
math_marks=marks[:3,1]
print(math_marks)
```

**Physics marks of first 4 students**

```python
phy_marks=marks[:4,2]
print(phy_marks)
```

**Overall Analysis**

```python
print(f"Maximum marks:{np.max(marks)}")
print(f"Mainimum marks:{np.min(marks)}")
print(f"Class average:{np.mean(marks)}")
```

```python
pass_mask=np.all(marks>=50,axis=1)
fail_mask=np.any(marks<50,axis=1)
pass_students=marks[pass_mask]
fail_students=marks[fail_mask]
print(f"pass_students:{pass_students}")
print(f"fail_students:{fail_students}")
```

```python
avg_students=np.mean(marks,axis=1)
grades=[]
#print(avg_students)
for i,avg in enumerate(avg_students):
  #print(f"Average marks of student{i+1}:{avg}")
  if avg>=80:
    grades.append(f"student{i}->Grade A")
  elif avg>=70 and avg<=84:
    grades.append(f"student{i}->Grade B")
  elif avg>=60 and avg<=69:
    grades.append(f"student{i}->Grade C")
  elif avg<60:
    grades.append(f"student{i}->Grade D")
  else:
    grades.append(f"student{i}->Grade F")
print(grades)
```

```python
avg_per_subject=np.mean(marks,axis=0)
print(avg_per_subject)
#[CS->75.4,Math->72.8,Physics77.4]
for i,avg in enumerate(avg_per_subject):
  lowest_avg=np.argmin(avg_per_subject)
if lowest_avg==0:
  print("CS is the toughest subject.")
elif lowest_avg==1:
  print("Math is the toughest subject.")
else:
  print("Physics is the toughest subject.")
```
